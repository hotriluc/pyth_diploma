```python
# Расчет параметра(длины ребра) сигнальной решетки - ребра "а"
a = np.sqrt((8/5)*Pw) 

# Случайный выбор сообщения (числа) для передачи по гауссовому каналу.
# Число должно находиться в диапазоне [0...(2^k - 1)]
x = np.floor(np.random.randint(0,np.power(2,k)-1))
print("Длина ребра \"а\" = ",a)
print("Случайное выбранное сообщение = ",x)

```
